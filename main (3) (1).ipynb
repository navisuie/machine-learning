```python
X=df.iloc[:,[0,1,2,3]].values 
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0) 


y_kmeans = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 30, c = 'yellow', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 30, c = 'red', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 30, c = 'black', label = 'Iris-virginica')


plt.title("Figure 2: Cluster of Iris")
plt.ylabel('Sepal_Width',color = "black")
plt.xlabel('Sepal_Length',color = "black")
plt.grid()
plt.legend()
plt.show()
```
